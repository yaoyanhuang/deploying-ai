```python
# Questions that can be easily answered given the article
easy_questions = [
    "What nationality was Ada Lovelace?",
    "What was an important finding from Lovelace's seventh note?",
]

# Questions the article touches on but does not fully answer
medium_questions = [
    # Article mentions her mother's rationalist push and her "poetical science" identity separately,
    # but never explains how one shaped or pushed back against the other
    "How did the tension between her mother's strictly rationalist upbringing and Ada's own 'poetical science' philosophy manifest in her work?",
    # Article acknowledges the half-siblings existed but says nothing about whether Ada
    # had any contact or relationship with them
    "Did Ada Lovelace have any relationship with her half-siblings?",
]

# Questions the article does not contain enough information to answer
difficult_questions = [
    # Article names the seventh note and calls it a program, but never describes the algorithm itself
    "What specific algorithm is described in Lovelace's seventh note and how does it work?",
    # The article ends at Lovelace's death in 1852; 20th-century influence is not discussed
    "How did Lovelace's ideas influence the inventors of the first electronic computers in the 1940s?",
]
```

We use our simple interface.

```python
def get_completion(
    input: list[dict[str, str]],
    model: str = MODEL,
    max_tokens=500,
    temperature=0,
    tools=None,
    logprobs=None,  # whether to return log probabilities of the output tokens or not. If true, returns the log probabilities of each output token returned in the content of message..
    top_logprobs=None,
) -> str:
    params = {
        "model": model,
        "input": input,
        "max_output_tokens": max_tokens,
        "temperature": temperature,
        "tools": tools,
        "include": ["message.output_text.logprobs"] if logprobs else [],
        "top_logprobs": top_logprobs,
    }
    if tools:
        params["tools"] = tools

    completion = client.responses.create(**params)
    return completion
```

The prompt below indicates that the model should determine if it has sufficient information to answer a question. We can assess the response to this query using logprobs.

```python
PROMPT = """You retrieved this article: {article}. The question is: {question}.
Before even answering the question, consider whether you have sufficient information in the article to answer the question fully.
Your output should JUST be the boolean true or false, indicating if you have sufficient information in the article to answer the question.
Respond with just one word, the boolean true or false. You must output the word 'True', or the word 'False', nothing else.
"""
```

```python

def has_sufficient_context_for_answer(article, question):
    output = ""
    API_RESPONSE = get_completion(
        [
            {
                "role": "user",
                "content": PROMPT.format(
                    article=article, question=question
                ),
            }
        ],
        model="gpt-4o-mini",
        logprobs=True,
    )
    output += f'\n\n**Question:** {question}\n\n'
    logprob = API_RESPONSE.output[0].content[0].logprobs[0]
    output += f'has_sufficient_context_for_answer: {logprob.token}, logprobs: {logprob.logprob}, linear probability: {np.round(np.exp(logprob.logprob)*100,2)}%\n'
    return(output)
```

### Easy Questions

```python

for qn in easy_questions:
    output = has_sufficient_context_for_answer(ada_lovelace_article, qn)
    display(Markdown(output))
```

### Medium Questions

```python
for qn in medium_questions:
    output = has_sufficient_context_for_answer(ada_lovelace_article, qn)
    display(Markdown(output))
```

### Difficult Questions

These questions go beyond what the article covers. We expect the model to return `False` with high confidence — low logprob uncertainty signals the RAG system should withhold an answer rather than risk hallucination.

```python
for qn in difficult_questions:
    output = has_sufficient_context_for_answer(ada_lovelace_article, qn)
    display(Markdown(output))
```
